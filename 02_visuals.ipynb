## Visualizations

In order to create visuals within this project, I had to conduct a lot of outside research and gather data. The sources in which this data came from can be found within the last slide of the deck in the Github Repository.

```python
# import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
```

The four countries listed alongside China below are the four countries with the highest physician to population ratios. Despite Portugal and Norway having around 2.5 times as many physicians as China, they too are facing a physician shortage. Physician shortages are an issue worldwide and it is estimated that by 2030 in a world that will need 80 millian physicians, we'll be short 18 million. Hiring more physicians to sooner detect pneumonia in children at Guangzhou Women and Children's Medical Center would not be a feasible option given that the supply needed does not exist.

```python
countries = {'Greece':6.1, 'Austria':5.2, 'Portugal':5.0, 'Norway':4.7, 'China':2.0}
```

```python
# plot Physician rate comparisons
x = range(0,6)
plt.figure(figsize=(10,8))
plt.bar(countries.keys(), countries.values(), width=.5, color=('darkred', 'darkred', 'darkred', 'darkred', 'salmon'))
plt.ylabel('Physicians Per 1,000 People', size=13)
plt.xlabel('Countries',  size=14)
plt.title('Countries With The Highest Physician To Population Ratios', size=14)

for index,data in enumerate(countries.values()):
    plt.text(x=index , 
             y =data+.1 , 
             s=f"{data}" , 
             fontdict=dict(fontsize=11)) # plot ratios over bars
fig2 = plt.gcf()
plt.show()
fig2.savefig('Visuals/doctor', dpi=100)
```

Again, thanks to outside research I learned that pneumonia is the leading cause of death in children under 5 across the globe, with over three quarters of a million children dying every year from the illness. The most concerning part about this number, is the fact that pneumonia is an entirely preventable illness. We have the treatments available, but these children are still dying. In China alone, 17% of all infantdeaths are due to pneumonia.

```python
cause = ['Pneumonia', 'Neonatal complications', 'Diarrheal diseases', 'Neonatal trauma', 'Malaria']
counts = [808920, 649439, 533768, 533250, 354294]
```

```python
# Plot the total crashes
plt.figure(figsize=(14,6))
sns.barplot(x=counts, y=cause, orient = 'h', palette='Blues_d')

for index,data in enumerate(counts):
    plt.text(y=index , 
             x=data+5 , 
             s=f"{data}" , 
             fontdict=dict(fontsize=12)) # plot percentages over bars

plt.title('Leading Causes of Death In Children Under 5 Worldwide', size=16)
    
fig1 = plt.gcf()
plt.show()
fig1.savefig('Visuals/cause', dpi=70)
    
plt.show()
```
