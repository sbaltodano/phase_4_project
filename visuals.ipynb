```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
```

```python
countries = {'Greece':6.1, 'Austria':5.2, 'Portugal':5.0, 'Norway':4.7, 'China':2.0}
```

```python
# plot churn comparisons
x = range(0,6)
plt.figure(figsize=(10,8))
plt.bar(countries.keys(), countries.values(), width=.5, color='darkred')
plt.ylabel('Physicians Per 1,000 People', size=13)
plt.xlabel('Countries',  size=14)
plt.title('Countries With The Highest Physician To Population Ratios', size=14)

for index,data in enumerate(countries.values()):
    plt.text(x=index , 
             y =data+.1 , 
             s=f"{data}" , 
             fontdict=dict(fontsize=11)) # plot percentages over bars
fig2 = plt.gcf()
plt.show()
fig2.savefig('Visuals/doctor', dpi=100)
```
